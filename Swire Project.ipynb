```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure consistent casing and strip any extra spaces
work_order_df['MAINTENANCE_ACTIVITY_TYPE'] = work_order_df['MAINTENANCE_ACTIVITY_TYPE'].str.strip().str.capitalize()

# Check unique values to confirm consistency
print(work_order_df['MAINTENANCE_ACTIVITY_TYPE'].unique())

# Plotting the count plot for Planned vs. Unplanned activities
plt.figure(figsize=(8, 6))
sns.countplot(data=work_order_df,
              x='MAINTENANCE_ACTIVITY_TYPE',
              order=['Planned', 'Unplanned'])  # Ensuring specific order if needed
plt.title('Count of Planned vs. Unplanned Maintenance Activities')
plt.xlabel('Maintenance Activity Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure date columns are in datetime format
work_order_df['EXECUTION_FINISH_DATE'] = pd.to_datetime(work_order_df['EXECUTION_FINISH_DATE'])  # Replace with the correct column name

# Convert actual work minutes to hours
work_order_df['ACTUAL_WORK_HOURS'] = work_order_df['ACTUAL_WORK_IN_MINUTES'] / 60  # Convert minutes to hours

# Extract year from the execution finish date
work_order_df['YEAR'] = work_order_df['EXECUTION_FINISH_DATE'].dt.year

# Get unique years to plot separate graphs for each year
unique_years = work_order_df['YEAR'].unique()

# Plotting the trend of actual work hours over time, separated by year
for year in unique_years:
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=work_order_df[work_order_df['YEAR'] == year], x='EXECUTION_FINISH_DATE', y='ACTUAL_WORK_HOURS')  # Corrected column name
    plt.title(f'Trend of Actual Work in Hours Over Time - {year}')
    plt.xlabel('Actual Finish Date')
    plt.ylabel('Actual Work Hours')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group the data by Production Location and Maintenance Activity Type, then count the occurrences
grouped_data = work_order_df.groupby(['PRODUCTION_LOCATION', 'MAINTENANCE_ACTIVITY_TYPE']).size().reset_index(name='Count')

# Calculate the total counts for each production location
total_counts = grouped_data.groupby('PRODUCTION_LOCATION')['Count'].transform('sum')

# Calculate the percentage of each maintenance activity type within each production location
grouped_data['Percentage'] = (grouped_data['Count'] / total_counts) * 100

# Pivot the data to create columns for each maintenance activity type with percentages
pivot_data = grouped_data.pivot(index='PRODUCTION_LOCATION', columns='MAINTENANCE_ACTIVITY_TYPE', values='Percentage').fillna(0)

# Define custom colors for the bars
custom_colors = ['green', 'red']

# Plotting the grouped bar plot with custom colors
ax = pivot_data.plot(kind='bar', figsize=(14, 6), color=custom_colors, width=0.8)
plt.title('Percentage of Planned vs Unplanned Maintenance Activities by Production Location')
plt.xlabel('Production Location')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Maintenance Activity Type', loc='upper right')

# Annotate each bar with percentage values
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center', fontsize=10, padding=3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Ensure the FUNCTIONAL_AREA_NODE_2_MODIFIED column is properly processed
# Group by Functional Area and Maintenance Activity Type, then count occurrences
grouped_data = work_order_df.groupby(['FUNCTIONAL_AREA_NODE_2_MODIFIED', 'MAINTENANCE_ACTIVITY_TYPE']).size().reset_index(name='Count')

# Pivot the data to have Maintenance Activity Types as columns
pivot_table = pd.pivot_table(grouped_data,
                             values='Count',
                             index='FUNCTIONAL_AREA_NODE_2_MODIFIED',
                             columns='MAINTENANCE_ACTIVITY_TYPE',
                             aggfunc='sum',
                             fill_value=0)

# Calculate percentages
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table['Planned_Percentage'] = (pivot_table.get('Planned', 0) / pivot_table['Total']) * 100
pivot_table['Unplanned_Percentage'] = (pivot_table.get('Unplanned', 0) / pivot_table['Total']) * 100

# Display the table with all rows
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns (if needed)

print(pivot_table[['Planned_Percentage', 'Unplanned_Percentage']])
```

```python
# Ensure the FUNCTIONAL_AREA_NODE_2_MODIFIED column is properly processed
# Group by Functional Area and Maintenance Activity Type, then calculate counts and mean of actual work minutes
grouped_data = work_order_df.groupby(['FUNCTIONAL_AREA_NODE_2_MODIFIED', 'MAINTENANCE_ACTIVITY_TYPE']).agg(
    Avg_Work_Minutes=('ACTUAL_WORK_IN_MINUTES', 'mean')
).reset_index()

# Pivot the data to have Maintenance Activity Types as columns
pivot_table = pd.pivot_table(grouped_data,
                             values='Avg_Work_Minutes',
                             index='FUNCTIONAL_AREA_NODE_2_MODIFIED',
                             columns='MAINTENANCE_ACTIVITY_TYPE',
                             aggfunc='mean',
                             fill_value=0)

# Rename columns for clarity
pivot_table.columns = [f'Avg_Work_Minutes_{col}' for col in pivot_table.columns]

# Round the values to two decimal places
pivot_table = pivot_table.round(2)

# Display the table with all rows
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns (if needed)

# Show the final table including the average work minutes
print(pivot_table)
```

```python
import pandas as pd

# Group by Production Location and Functional Area Node 2, calculating counts and mean of actual work minutes
grouped_data = work_order_df.groupby(['PRODUCTION_LOCATION', 'FUNCTIONAL_AREA_NODE_2_MODIFIED']).agg(
    Count=('ACTUAL_WORK_IN_MINUTES', 'size'),
    Avg_Work_Minutes=('ACTUAL_WORK_IN_MINUTES', 'mean')
).reset_index()

# Filter out rows where the average work minutes are 0.00
grouped_data = grouped_data[grouped_data['Avg_Work_Minutes'] != 0.00]

# Calculate the total counts per Production Location
total_counts = grouped_data.groupby('PRODUCTION_LOCATION')['Count'].sum().reset_index()
total_counts = total_counts.rename(columns={'Count': 'Total_Count'})

# Merge total counts back into the grouped data
grouped_data = pd.merge(grouped_data, total_counts, on='PRODUCTION_LOCATION')

# Calculate the percentage
grouped_data['Percentage'] = (grouped_data['Count'] / grouped_data['Total_Count']) * 100

# Pivot the data to have Functional Area Node 2 Modified as columns
pivot_table = pd.pivot_table(
    grouped_data,
    values='Percentage',
    index='PRODUCTION_LOCATION',
    columns='FUNCTIONAL_AREA_NODE_2_MODIFIED',
    aggfunc='sum',
    fill_value=0
)

# Flatten MultiIndex columns
pivot_table.columns = [f'{col}' for col in pivot_table.columns]

# Round the values to two decimal places
pivot_table = pivot_table.round(2)

# Display the table with all rows
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns (if needed)

# Show the final table including the percentages
print(pivot_table)
```
